```python
X= df.drop('quality', axis=1)
Y= df['quality']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
from sklearn import svm
```

```python
model_1 = svm.SVC(kernel="rbf")
model_1.fit(X_train, Y_train)
Y_pred = model_1.predict(X_test) 
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix
print('MSE : \t', mean_squared_error(Y_test, Y_pred))
print('R2_score : \t\t', model_1.score(X_train, Y_train))
print('Accuracy score : \t', accuracy_score(Y_test, Y_pred))
print('Confusion Matrix: \n', confusion_matrix(Y_test, Y_pred))
```

```python
model_2 = svm.SVC(kernel="linear")
model_2.fit(X_train,Y_train)
Y_pred = model_2.predict(X_test)
```

```python
print('MSE : \t', mean_squared_error(Y_test, Y_pred))
print('R2_score : \t\t', model_2.score(X_train, Y_train))
print('Accuracy score : \t', accuracy_score(Y_test, Y_pred))
print('Confusion Matrix: \n', confusion_matrix(Y_test, Y_pred))
```
